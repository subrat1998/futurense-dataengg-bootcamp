Reading The Bank Marketing Dataset

```
from pyspark.sql.functions import *
df=spark.read.csv("/FileStore/tables/bankmarketdata.csv",sep=';',header=True,inferSchema=True)
```

```
df.show(5)
```

Marketing Success Rate Analysis

```
(df.filter(df.y == 'yes').count()/df.count())*100
```

Marketing Failure Rate Analysis

```
(df.filter(df.y == 'no').count()/df.count())*100
```

Targeted Customer Analysis

```
df.select(\
      min('age').alias('min_age'),\
      max('age').alias('max_age'),\
      round(mean('age')).alias('avg_age')\
      ).show()
```

Quality Of Customers Analysis

```
df.select(\
      mean('balance').alias('avg_balance'),\
      median('balance').alias('median_balance'),\
      ).show()
```

Age Wise Marketing Subscription Analysis

```
df.filter(df.y == 'yes').groupby('age').count().show()
```

```
age_expr = when(df.age.between(13, 19), "Teenagers") \
            .when(df.age.between(20, 40), "Youngsters") \
            .when(df.age.between(40, 60), "MiddleAgers") \
            .otherwise("Seniors")

age_count_df = df.filter(df.y == 'yes').groupby(age_expr.alias("age_group")).agg(count("y").alias("subscription_count"))
age_count_df.show()
```

Maritial Status Wise Marketing Subscription Analysis

```
df.filter(df.y == 'yes').groupby('marital').count().show()
```

Age-Maritial Status Wise Marketing Subscription Analysis

```
df.filter(df.y == 'yes').groupby('age','marital').count().show()
```
